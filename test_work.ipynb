```python
X = df_c[['Общая площадь', 'Расстояние до центра, м.']]
y = df_c['Удельная стоимость']
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X, y)
y_pred = slr.predict(X)
```

```python
fig = plt.figure(figsize=(10,7))
sns.regplot(x=df_c[['Расстояние до центра, м.']], y=df_c['Удельная стоимость'], color='blue', marker='+')
```

```python
sns.regplot(x=df_c[['Общая площадь']], y=df_c['Удельная стоимость'], color='magenta', marker='+')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rmse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print('Slope:' ,slr.coef_)
print('Intercept:', slr.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```
