```python
import numpy as np

INPUT_DIM = 784 #количество входных узлов 
OUTPUT_DIM = 10 #количество выходных узлов
H_DIM = 100 #количество скрытых узлов
learning_index = 0.3 #коэффициент обучения
epochs = 5

W1 = np.random.random((H_DIM, INPUT_DIM)) - 0.5
W2 = np.random.random((OUTPUT_DIM, H_DIM)) - 0.5

def sigmoid(x):
    return 1/(1+np.exp(-x))

def forward(W1, W2, input_list):
    
    hidden_input_x = W1 @ input_list 
    hidden_output_x = sigmoid(hidden_input_x)
    
    input_y = W2 @ hidden_output_x 
    output_y = sigmoid(input_y)
    
    return (output_y, hidden_output_x)

def training(W1, W2, input_list, real_list):
    
    x = np.array(input_list, ndmin=2).T
    real = np.array(real_list, ndmin=2).T
    output_y, hidden_output_x = forward(W1, W2, x)
    out_errors = real - output_y
    hidden_errors = W2.T @ out_errors
    W2 += learning_index * (out_errors * output_y * (1.0 - output_y)) @ (hidden_output_x.T)
    W1 += learning_index * (hidden_output_x * hidden_errors * (1 - hidden_output_x)) @ (x.T)  

def query(inputs_list):
    inputs = np.array(inputs_list, ndmin=2).T
    hidden_inputs = W1 @ inputs_list
    hidden_outputs = sigmoid(hidden_inputs)
    final_inputs = W2 @ hidden_outputs
    final_outputs = sigmoid(final_inputs)
    return final_outputs

def begin_training(file):
    scorecard = []
    with open(file,'r') as train_data_file:
        train_data_list = train_data_file.readlines()
        train_data_file.close()
    for e in range(epochs):
        for i in train_data_list:
            all_values = i.split(',')
            inputs = (np.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
            targets = np.zeros(OUTPUT_DIM) + 0.01
            targets[int(all_values[0])] = 0.99
            training(W1, W2, inputs, targets)

def begin_testing(file):
    scorecard = []
    with open(file,'r') as train_data_file:
        train_data_list = train_data_file.readlines()
        train_data_file.close()
    scorecard_test = []
    for record in train_data_list:
        label = 0
        all_values = record.split(',')
        correct_label = int(all_values[0])
        inputs = (np.asfarray(all_values[1:])/255.0*0.99) + 0.1
        outputs = query(inputs)
        label = np.argmax(outputs)
        if (label == correct_label):
            scorecard.append(1)
        else:
            scorecard.append(0)
        print('Верная цифра = ', correct_label)
        print('Вывод нейросети = ', label)
    scorecard_array = np.asarray(scorecard)
    print('Обучение завершено: Эффективность = ', np.round(scorecard_array.sum()*100/scorecard_array.size), '%')
    
begin_training('mnist_train.csv')
begin_testing('mnist_test.csv')
```
