````python
import re
import pandas as pd
res = ''
name = []
all_citizens = citizenA + citizenB + citizenC
all_citizens = all_citizens.split(', ')
all_citizens_text = ''.join(all_citizens)
res = re.findall(r'[A-Z][a-z]+', all_citizens_text) 
for word in res:
    if word != 'Female' and word != 'Male':
        name.append(word)
print(name)

gender = []
res = re.findall(r'[A-Z][a-z]+', all_citizens_text) 
for word in res:
    if word == 'Female' or word == 'Male':
        gender.append(word)
print(gender)

height = []
res = re.findall(r'[\d\.]+\scm', all_citizens_text)  
for word in res:
    height.append(int(round(float(word[:-3]))))
print(height)

weight = []
res = re.findall(r'[\d\.]+\skg', all_citizens_text) 
for word in res:
    weight.append(int(round(float(word[:-3]))))
print(weight)

age = []
res = re.findall(r'[\d\.]+\syears', all_citizens_text) 
for word in res:
    age.append(int(round(float(word[:-6]))))
print(age)

indexes = []
for one_height in height:
    if one_height > 170:
        indexes.append(height.index(one_height))

def high(data, var): 
    for i, v in enumerate(data):
        if i in indexes:
            var.append(v)
    return var

name_high = []
high(name, name_high)
print(name_high)
gender_high = []
high(gender, gender_high)
print(gender_high)
height_high = []
high(height, height_high)
print(height_high)
weight_high = []
high(weight, weight_high)
print(weight_high)
age_high = []
high(age, age_high)
print(age_high)

research_high = pd.DataFrame({
    'name_high': name_high, 
    'gender_high': gender_high, 
    'height_high': height_high, 
    'weight_high': weight_high,
    'age_high': age_high
})
print(research_high)
````

You have a number of texts and a list of "special" words. Your task is to replace all the words in the texts from the list of special words with one of these characters: $*&%#@$

Write a function **replace_words** that accepts as input:
* **at least** one text 
* a list of "special" words (it is guaranteed that all words in the list start with a lowercase letter)
* the sign to replace the words with

At the output, the function should give:
* a list with the original texts in which the "special" words are replaced with a character from the input

The number of characters must correspond to the number of letters in the replaced word.

Words with uppercase and lowercase letters are considered the same (funny == Funny)

Example:

<blockquote>
text7 = 'Numbing the pain for a while will make it worse when you finally feel it.'
    
text8 = 'It is a strange thing, but when you are dreading something, and would give anything to slow down time, it has a disobliging habit of speeding up.'
    
words = ['pain', 'it', 'thing', 'disobliging']

replace_words(text7, words, '#')
    <blockquote>
       ['Numbing the #### for a while will make ## worse when you finally feel ##.']
    </blockquote>
replace_words(text7, words, '%', text8)
    <blockquote>
      ['Numbing the %%%% for a while will make %% worse when you finally feel %%.',
 '%% is a strange %%%%%, but when you are dreading something, and would give anything to slow down time, %% has a %%%%%%%%%%% habit of speeding up.']
     </blockquote>   
</blockquote>

````python
import re

text = '''A computer programmer, sometimes called a software developer,
a programmer or more recently a coder (especially in more informal contexts),
is a person who creates computer software. The term computer programmer
can refer to a specialist in one area of computers or to a generalist
who writes code for many kinds of software.'''

text2 = '''
There is no industry-wide standard terminology, so "programmer" and 
"software engineer" might refer to the same role at different companies. 
Most typically, someone with a job title of "programmer" or "software developer" 
might focus on implementing a detailed specification into computer code, 
fixing bugs, and performing code reviews. They might have a degree in computer 
science, an associate degree, or might be self-taught or attended a programming boot camp.'''

words = ['developer', 'code', 'coder', 'programmer']

def replace_words(text, words, symbol, *argv):
    text_split = re.split('\s|,|\.|"|!|\?', text)
    for word in text_split:
        if word in words:
            text = text.replace(word, len(word)*symbol)
    if argv:
        sentences = [text]
        for arg in argv:
            text_split = re.split('\s|,|\.|"|"', arg)
            for word in text_split:
                if word in words:
                    arg = arg.replace(word, len(word)*symbol)
            sentences.append(arg)
        return sentences
    else:
        return [text]
        
print(replace_words(text, words, '%', text2))
````
